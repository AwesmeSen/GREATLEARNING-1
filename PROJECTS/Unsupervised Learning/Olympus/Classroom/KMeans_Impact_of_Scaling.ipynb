```python
import numpy as np
import seaborn
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

rnorm = np.random.randn

x = rnorm(1000) * 10  
y = np.concatenate([rnorm(500), rnorm(500) + 5])

Figure = plt.figure(figsize=(20,20))

#fig, axes = Figure.subplots(3,1)
axes = Figure.subplots(3,1)

axes[0].scatter(x, y)
axes[0].set_title('Data (note different axes scales)')

km = KMeans(2)

clusters = km.fit_predict(np.array([x, y]).T)   # no scaling which means X axis will have higher contribution to distance
axes[1].scatter(x, y, c=clusters, cmap='bwr')
axes[1].set_title('non-normalised K-means')



clusters = km.fit_predict(np.array([x/10, y/2]).T)  # Scaling for clustering
axes[2].scatter(x/10, y/2, c=clusters, cmap='bwr')  # plotting the new scale on x and y
axes[2].set_title('Normalised K-means')

```

```python
import numpy as np
import seaborn
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy.stats import zscore

rnorm = np.random.randn

x = np.concatenate([rnorm(500)*10, rnorm(500)*10 +30 ])
y = np.concatenate([rnorm(500)*5, rnorm(500)*5 + 10])

x_z = zscore(x)
y_z = zscore(y)

Figure = plt.figure(figsize=(20,20))

#fig, axes = Figure.subplots(3,1)
axes = Figure.subplots(3,1)

axes[0].scatter(x, y)
axes[0].set_title('Data (note different axes scales)')

km = KMeans(2)

clusters = km.fit_predict(np.array([x, y]).T)   # no scaling which means X axis will have higher contribution to distance
axes[1].scatter(x, y, c=clusters, cmap='bwr')
axes[1].set_title('non-normalised K-means')



clusters = km.fit_predict(np.array([x_z, y_z]).T)  # Scaling for clustering
axes[2].scatter(x_z, y_z, c=clusters, cmap='bwr')  # plotting the new scale on x and y
axes[2].set_title('Normalised K-means')

# Repeating the same step again to prevent color switch between clusters. Did not know how to prevent the color switch that 
# one can notice if we do not do the code below

clusters = km.fit_predict(np.array([x_z, y_z]).T)  # Scaling for clustering
axes[2].scatter(x_z, y_z, c=clusters, cmap='bwr')  # plotting the new scale on x and y
axes[2].set_title('Normalised K-means')
```
